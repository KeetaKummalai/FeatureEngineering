```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.linear_model import LinearRegression
```

##create x variable with 200 observation and normal distribution

```python
x = np.random.randn(200)
plt.hist(x, bins=50)
```

```python
y = x * 10 + np.random.rand(200) * 2
```

```python
df = pd.DataFrame([x,y])
df1 = pd.DataFrame([x,y]).T #.T transpote
```

```python
df1.columns = ['x', 'y']
```

```python
sns.lmplot(x='x', y='y', data=df1, order =1)
plt.ylabel('Target')
plt.xlabel('Independent Variable')
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(df1['x'].to_frame(), df1['y'])
```

```python
predictions = linreg.predict(df1['x'].to_frame())
residuals = df1['y'] -predictions
plt.scatter(y=residuals, x=df1['x'])
plt.ylabel('Resifuals')
plt.xlabel('Indepent variables x')
```

```python
sns.displot(residuals, bins=30)
plt.xlabel('Residuals')
```
