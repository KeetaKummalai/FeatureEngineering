```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cols = ['GENDER', 'RFA_2', 'MDMAUD_A', 'RFA_2', 'DOMAIN', 'RFA_15']
df = pd.read_csv('../cup98LRN.txt', usecols=cols)
```

```python
data = df.replace('',np.nan)
data.head()
```

```python
data.nunique(dropna=False)
```

```python
data['GENDER'].unique()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../car.txt', header=None)
df.columns = ['buying', 'maint','doors', 'persons', 'lug_booy','safety','class']
df.head()
```

```python
df['class'].unique()
```

```python
label_freq = df['class'].value_counts()/len(df)
print(label_freq)
```

```python
fig = label_freq.sort_values(ascending=False).plot.bar()
fig.axhline(y=0.05, color='red')
fig.set_ylabel('percentage of cars within each category')
fig.set_xlabel('Variable: class')
fig.set_title('Identifying Rare')
plt.show()
```
